# Paired Student’s t-test

Tests whether the means of two paired samples are significantly different.

Assumptions

Observations in each sample are independent and identically distributed (iid).
Observations in each sample are normally distributed.
Observations in each sample have the same variance.
Observations across each sample are paired.
Interpretation

H0: the means of the samples are equal.

H1: the means of the samples are unequal.

```python
# Paired Student's t-test
from scipy.stats import ttest_rel
test_params = ['wpm', 'acc', 'rawWpm', 'consistency', 'correct', 'extra', 'missed']
for param in test_params:
  print(param)
  no_music = df[param].to_numpy()
  white = df_white[param].to_numpy()
  music = df_music[param].to_numpy()
  incr = df_incr[param].to_numpy()
  # no_music v/s white, no_music v/s music, no_music v/s incr, white v/s music,
  # white v/s incr, music v/s incr
  stat, p = ttest_rel(no_music, white)
  print('no_music v/s', 'white')
  print('stat=%.3f, p=%.3f' % (stat, p))
  if p > 0.05:
    print('Probably the same distribution')
  else:
    print('Probably different distributions')
  stat, p = ttest_rel(no_music, music)
  print('no_music v/s', 'music')
  print('stat=%.3f, p=%.3f' % (stat, p))
  if p > 0.05:
    print('Probably the same distribution')
  else:
    print('Probably different distributions')
  stat, p = ttest_rel(no_music, incr)
  print('no_music v/s', 'incr')
  print('stat=%.3f, p=%.3f' % (stat, p))
  if p > 0.05:
    print('Probably the same distribution')
  else:
    print('Probably different distributions')
  stat, p = ttest_rel(white, music)
  print('white v/s', 'music')
  print('stat=%.3f, p=%.3f' % (stat, p))
  if p > 0.05:
    print('Probably the same distribution')
  else:
    print('Probably different distributions')
  stat, p = ttest_rel(white, incr)
  print('white v/s', 'incr')
  print('stat=%.3f, p=%.3f' % (stat, p))
  if p > 0.05:
    print('Probably the same distribution')
  else:
    print('Probably different distributions')
  stat, p = ttest_rel(music, incr)
  print('music v/s', 'incr')
  print('stat=%.3f, p=%.3f' % (stat, p))
  if p > 0.05:
    print('Probably the same distribution')
  else:
    print('Probably different distributions')
  print()
```

# Wilcoxon's Signed Rank Test

Tests whether the distributions of two paired samples are equal or not.

Assumptions

Observations in each sample are independent and identically distributed (iid).
Observations in each sample can be ranked.
Observations across each sample are paired.
Interpretation

H0: the distributions of both samples are equal.
H1: the distributions of both samples are not equal.

```python
# Wilcoxon's Signed Rank Test
from scipy.stats import wilcoxon
test_params = ['wpm', 'acc', 'rawWpm', 'consistency', 'correct', 'extra', 'missed']
for param in test_params:
  print(param)
  no_music = df[param].to_numpy()
  white = df_white[param].to_numpy()
  music = df_music[param].to_numpy()
  incr = df_incr[param].to_numpy()
  # no_music v/s white, no_music v/s music, no_music v/s incr, white v/s music,
  # white v/s incr, music v/s incr
  stat, p = wilcoxon(no_music, white)
  print('no_music v/s', 'white')
  print('stat=%.3f, p=%.3f' % (stat, p))
  if p > 0.05:
    print('Probably the same distribution')
  else:
    print('Probably different distributions')
  stat, p = wilcoxon(no_music, music)
  print('no_music v/s', 'music')
  print('stat=%.3f, p=%.3f' % (stat, p))
  if p > 0.05:
    print('Probably the same distribution')
  else:
    print('Probably different distributions')
  stat, p = wilcoxon(no_music, incr)
  print('no_music v/s', 'incr')
  print('stat=%.3f, p=%.3f' % (stat, p))
  if p > 0.05:
    print('Probably the same distribution')
  else:
    print('Probably different distributions')
  stat, p = wilcoxon(white, music)
  print('white v/s', 'music')
  print('stat=%.3f, p=%.3f' % (stat, p))
  if p > 0.05:
    print('Probably the same distribution')
  else:
    print('Probably different distributions')
  stat, p = wilcoxon(white, incr)
  print('white v/s', 'incr')
  print('stat=%.3f, p=%.3f' % (stat, p))
  if p > 0.05:
    print('Probably the same distribution')
  else:
    print('Probably different distributions')
  stat, p = wilcoxon(music, incr)
  print('music v/s', 'incr')
  print('stat=%.3f, p=%.3f' % (stat, p))
  if p > 0.05:
    print('Probably the same distribution')
  else:
    print('Probably different distributions')
  print()
```
